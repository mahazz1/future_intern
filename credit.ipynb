```python
df.drop(labels=["first","last","job","dob","trans_num","street","trans_date_trans_time","city","state"],axis=1,inplace=True)
```

```python
encoder=LabelEncoder()
new_col=encoder.fit_transform(df["merchant"].values)
df["merchant_new"]=new_col
df.drop(labels=["merchant"],axis=1,inplace=True)
```

```python
encoder=LabelEncoder()
new_col=encoder.fit_transform(df["category"].values)
df["category_new"]=new_col
df.drop(labels=["category"],axis=1,inplace=True)
```

```python
df=pd.get_dummies(df)
df.drop(labels=["gender_F"],axis=1,inplace=True)
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(),
            cmap = 'BrBG',
            fmt = '.2f',
            linewidths = 2,
            annot = True)
```

```python
correlation_matrix=df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix,cmap='coolwarm',annot=False,fmt=".2f")
plt.title("correlation_matrix")
plt.show()
```

```python
x=df.drop("is_fraud",axis=1)
y=df["is_fraud"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python

model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predict=model.predict(x_test)
predict
```

```python
print(accuracy_score(predict,y_test))
```
